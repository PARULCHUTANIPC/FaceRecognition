```python
import face_recognition
import os
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
ALLOWED_EXTENSIONS = set(['png', 'jpg', 'jpeg'])

def allowed_file(filename):
  return '.' in filename and filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS
```

```python
ref_encodings={}
for i in os.listdir('/content/drive/MyDrive/FR/refimages'):
    ref_name = i.split('.')[0].upper()
    ref_image = face_recognition.load_image_file('/content/drive/MyDrive/FR/refimages/{}'.format(i))
    ref_image_locations = face_recognition.face_locations(ref_image, model="cnn")
    ref_image_encoding = face_recognition.face_encodings(ref_image, ref_image_locations)[0]
    print(ref_image.shape)
    ref_encodings[ref_name] = ref_image_encoding
```

```python
def target_dir(key):
  parent_dir = "/content/drive/MyDrive/FR"
  path = os.path.join(parent_dir, key)

  if key in os.listdir('/content/drive/MyDrive/FR'):
    pass
  else:
    os.mkdir(path)

  return path
```

```python
for image in os.listdir('/content/drive/MyDrive/FR/Main'):
  if image and allowed_file(image):
    print(image)
    group_test_image = face_recognition.load_image_file('/content/drive/MyDrive/FR/Main/{}'.format(image))
    group_test_image_location = face_recognition.face_locations(group_test_image)
    print(group_test_image_location)
    
    for i in group_test_image_location:
      img = cv2.imread('/content/drive/MyDrive/FR/Main/AJ0A9451.JPG')
      (top, right, bottom, left) = i
      crop_img = group_test_image[top-50:bottom+50, left-50:right+50]
      print('cropped image - ', crop_img)
      if len(crop_img) >0:
        resize_crop_img = cv2.resize(crop_img, (100,100))
        cv2_imshow(resize_crop_img)
        print('****************************************************')
        crop_face_encoding = face_recognition.face_encodings(resize_crop_img)
        print(crop_face_encoding)

        if len(crop_face_encoding)!=0:
          for key in ref_encodings.keys():
            ref_encoding = ref_encodings[key]
            path = target_dir(key)
            faceDis = face_recognition.face_distance(ref_encoding, crop_face_encoding)
            if faceDis > 0.5:
              pass
            else:
              os.chdir(path)
              cv2.imwrite(image, group_test_image)
        else:
          pass
```
